```python
from sentence_transformers import SentenceTransformer, util
import json
with open("yenihaberler.json","r",encoding="UTF-8") as f: # get the data 
  ds=json.load(f)
  A_haber ="KamuBulteni" # a News
  haberler =ds.keys() # Take every keys
  sentences=[] # A news  sentences for compare 
  
  A=ds["KamuBulteni"].values()
  for n in A:
    sentences.append(n) # take from A news sentences to sentences list

  veriler1={}
  
#   sentences2=ds["Altinok"].values()
  for x in haberler:
    
    veriler=[]
    sentences2=[]
    if x!=A_haber:
      B=ds[x].values()
      for m in B:
         sentences2.append(m)
     
           model = SentenceTransformer('emrecan/bert-base-turkish-cased-mean-nli-stsb-tr') #Türkçe model farklı modelleride kullanabilirsiniz

      embeddings = model.encode(sentences)#Convert to vektor
      embeddings2 = model.encode(sentences2)

    
      embeddings.shape

      cosine_scores = util.cos_sim(embeddings, embeddings2) #cosine similarity 
      
      for i in range(len(sentences)): #x1 compare a news one by one from A  to All B news  sentences 
        high_score=[]
        for j in range(len(sentences2)): #y1....n
            
                 high_score.append({A_haber:sentences[i],  x:sentences2[j], "score" : float(cosine_scores[i][j]) })
                 
        max_age = 0
        # sava the max number to high_score dic
        for dict in high_score:
            if dict["score"] > max_age:
                max_age = dict["score"]
        
        
        for n in high_score  :
          if  n["score"] == max_age:
              veriler.append(n)
        
      
      veri={x:veriler}
      
      veriler1.update(veri)
 #after compare finished sava the data to json file By A news name
  with open("KamuBulteni.json","a",encoding="utf-8") as f:
      f.write(json.dumps(veriler1,ensure_ascii=False,indent=2))
    
```
